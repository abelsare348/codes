```
cam2_df=pd.read_json("/content/20240825T0000_DBA data.json")
```

```
cam2_df_a=json_normalize(cam2_df.get('Data DBA data',[]))
```

```
def flatten_nested_data(data):
    flattened = []
    for item in data:
        if isinstance(item, list):  # If item is a list, recursively process it
            flattened.extend(flatten_nested_data(item))
        elif isinstance(item, dict):  # If item is a dictionary, process it
            flattened.append(item)
    return flattened
```

```
cam2_df_c=cam2_df.get('Data DBA data',[])   # fetched the data in list to flattened
```

```
flattened_list = flatten_nested_data(cam2_df_c)
```

```
df = pd.json_normalize(flattened_list, sep='_')
```

```
if 'locations' in df.columns:
    df_locations = pd.json_normalize(df['locations'].explode(), sep='_')
    df = df.drop(columns=['locations']).join(df_locations)

if 'values' in df.columns:
    # Flatten 'values' if it's a list inside the dataframe
    df_values = df['values'].apply(lambda x: x[0] if isinstance(x, list) and len(x) > 0 else None)
    df['values'] = df_values

# Convert 'timestamp' to datetime
if 'timestamp' in df.columns:
    df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```
cam2_df_d=df.assign(file_name='Data DBA data')  # reassinged the data to cam_2_df
```

```
print(cam2_df_d.to_string())
```
